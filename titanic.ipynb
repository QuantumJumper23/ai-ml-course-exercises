```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head()
```

```python
columns_to_drop = ['PassengerId', 'Name', 'SibSp', 'Parch', 'Embarked','Ticket']
df.drop(columns=columns_to_drop, axis='columns', inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df['HasCabin']=df['Cabin'].notnull().astype(int)
df.drop(['Cabin'],axis='columns',inplace=True,errors='ignore')
df['Sex']=df['Sex'].map({'female':0,'male':1})
```

```python
df.head()
```

```python

happened=df[['Pclass','Sex','Age','Fare','HasCabin']].copy()
target=df['Survived']
print(happened.isnull().sum())
```

```python
#there are nan in age so we have to fix it i thought the kaggle data don't have nan but i was wrong 
#i have to fill it with mean
happened['Age'] =happened['Age'].fillna(happened['Age'].mean())
```

```python
from sklearn.naive_bayes import GaussianNB
model =GaussianNB()
X_train=happened
Y_train=target
model.fit(X_train,Y_train)
```

```python
#pclass,sex,age,fare,hascabin
ram_data=pd.DataFrame([[3,1,25,1,0]],columns=happened.columns)
prediction=model.predict(ram_data)
if prediction[0]==1:
    print("survived")
else:
    print("not survived")
```

```python
#Testing phase now as i have to find how well my model is
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(happened,target,test_size=0.2)
model.fit(x_train,y_train)
prediction=model.predict(x_test)
from sklearn.metrics import accuracy_score
print("Accuracy",str(int((accuracy_score(y_test,prediction))*100))+"%")
```
